# Аномиалии в данных

```python
from sklearn.ensemble import IsolationForest

# Удаление ненужных столбцов, если требуется
# Например, если столбцы "Country Name", "Country Code" не нужны для обнаружения аномалий
data_anomaly = data.drop(['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], axis=1)
```

```python
data_anomaly.info()
```

```python
# Заполнение пропущенных значений, если необходимо
# Например, можно заполнить пропуски средними значениями по столбцам
data_anomaly = data_anomaly.fillna(data_anomaly.mean())

# Создание модели Isolation Forest
model = IsolationForest(random_state=42)

# Обучение модели
model.fit(data_anomaly)

# Предсказание аномалий
anomaly_labels = model.predict(data_anomaly)

# Добавление меток аномалий к исходным данным
data_anomaly['Anomaly'] = anomaly_labels

# Вывод данных, включая метки аномалий
data_anomaly
```

# Перевернем данные

```python
import re

data['Indicator Name'] = data['Indicator Name'].apply(lambda x: re.split(r'\s\(', x)[0])
data['Indicator Name'] = data['Indicator Name'].apply(lambda x: x.split(' - ')[0])

data['Indicator Name']
```

```python
# Получим список всех уникальных показателей
indicator_names = data['Indicator Name'].unique()

# Создадим пустой DataFrame для объединения результатов
result = pd.DataFrame()

# Пройдемся по каждому показателю
for indicator_name in indicator_names:
    # Отфильтруем данные по текущему показателю
    filtered_data = data[data['Indicator Name'] == indicator_name]

    # Удалим ненужные столбцы
    filtered_data = filtered_data.drop(columns=['Country Name', 'Country Code', 'Indicator Code'])

    # Преобразуем данные с помощью pivot_table
    pivot_data = filtered_data.pivot_table(index=None, columns='Indicator Name', aggfunc='sum')

    # Переименуем столбцы, чтобы они были уникальными
    pivot_data.columns = [f"{indicator_name} - {col}" for col in pivot_data.columns]

    # Объединим результат с общим DataFrame
    result = pd.concat([result, pivot_data], axis=1)
```

```python
result.columns = result.columns.str.split(' - ').str[0]
result.head()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from catboost import CatBoostRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

X_train, X_test, y_train, y_test = train_test_split(result.drop(['GDP'], axis=1), result['GDP'], test_size=0.2, random_state=42)
```

```python
# Инициализация и обучение модели
ctb = CatBoostRegressor(iterations=1000, learning_rate=0.1, depth=6, loss_function='RMSE', random_seed=42)
ctb.fit(X_train, y_train, eval_set=(X_test, y_test), verbose=0, plot=True)
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Инициализация и обучение моделей
models = {
    "Random Forest": RandomForestRegressor(random_state=42),
    "Gradient Boosting": GradientBoostingRegressor(random_state=42),
    "Linear Regression": LinearRegression(),
    "SVR": SVR()
}

results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    rmse = mean_squared_error(y_test, y_pred, squared=False)
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    results[name] = {"RMSE": rmse, "MSE": mse, "MAE": mae, "R2": r2}

y_pred_catboost = ctb.predict(X_test)
rmse_catboost = mean_squared_error(y_test, y_pred_catboost, squared=False)
mse_catboost = mean_squared_error(y_test, y_pred_catboost)
mae_catboost = mean_absolute_error(y_test, y_pred_catboost)
r2_catboost = r2_score(y_test, y_pred_catboost)

# Добавление результатов для CatBoostRegressor в словарь
results["CatBoost"] = {"RMSE": rmse_catboost, "MSE": mse_catboost, "MAE": mae_catboost, "R2": r2_catboost}

metrics_df = pd.DataFrame(results)
metrics_df
```

# Прогнозирование с помощью временных рядов

```python
result_df = result['GDP']
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Здесь должен быть ваш DataFrame result_df

# Масштабирование данных
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(np.array(result_df).reshape(-1, 1))

# Создание обучающих данных
def create_dataset(data, time_step):
    X, y = [], []
    for i in range(len(data) - time_step):
        X.append(data[i:(i + time_step), 0])
        y.append(data[i + time_step, 0])
    return np.array(X), np.array(y)

time_step = 12  # количество месяцев для использования в качестве временных шагов
X_train, y_train = create_dataset(scaled_data, time_step)

# Преобразование данных в формат, пригодный для LSTM
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)

# Инициализация модели LSTM
model_lstm = Sequential()
model_lstm.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model_lstm.add(LSTM(units=50))
model_lstm.add(Dense(1))

# Компиляция модели
model_lstm.compile(optimizer='adam', loss='mean_squared_error')

# Обучение модели
model_lstm.fit(X_train, y_train, epochs=100, batch_size=1)

# Создание тестовых данных
test_data = scaled_data[-time_step:]
test_data = test_data.reshape(1, -1)

# Прогнозирование
X_test = np.array(test_data)
X_test = np.reshape(X_test, (1, time_step, 1))
predicted_GDP_scaled = model_lstm.predict(X_test)
predicted_GDP = scaler.inverse_transform(predicted_GDP_scaled)
```

```python
# Прогнозирование на 2020, 2021 и 2022 годы
predictions = []

for i in range(3):
    X_test_reshaped = np.reshape(X_test, (1, time_step, 1))
    predicted_GDP_scaled = model_lstm.predict(X_test_reshaped)
    predicted_GDP = scaler.inverse_transform(predicted_GDP_scaled)
    predictions.append(predicted_GDP[0][0])

    # Обновление X_test для следующего прогноза
    X_test = np.concatenate((X_test[:, 1:, :], predicted_GDP_scaled.reshape(1, 1, 1)), axis=1)

# Получение реальных значений GDP для 2020, 2021 и 2022 годов из исходного DataFrame
real_values = result_df[-3:].values

# Визуализация прогноза
plt.plot(result_df.index, result_df.values, label='Исходные данные')
plt.plot(range(2020, 2023), predictions, label='Прогноз', color='yellow')
plt.plot(range(2020, 2023), real_values, label='Реальные значения', color='red', linestyle='--')
plt.xlabel('Год')
plt.ylabel('GDP')
plt.title('Прогноз GDP с помощью LSTM')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

# Реальные значения GDP за последние три года
real_values = result_df[-3:].values

# Прогнозные значения GDP за последние три года
predicted_values = predictions[-3:]

# Рассчитываем RMSE
rmse = sqrt(mean_squared_error(real_values, predicted_values))
print("RMSE для последних трех лет прогноза:", rmse)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from math import sqrt

# Реальные значения GDP за последние три года
real_values = result_df[-3:].values

# Прогнозные значения GDP за последние три года
predicted_values = predictions[-3:]

# Рассчитываем MSE
mse = mean_squared_error(real_values, predicted_values)
print("MSE для последних трех лет прогноза:", mse)

# Рассчитываем RMSE
rmse = sqrt(mse)
print("RMSE для последних трех лет прогноза:", rmse)

# Рассчитываем R2
r2 = r2_score(real_values, predicted_values)
print("R2 для последних трех лет прогноза:", r2)

# Рассчитываем MAE
mae = mean_absolute_error(real_values, predicted_values)
print("MAE для последних трех лет прогноза:", mae)
```

# Сранвнение всех моделей

```python
# Добавление метрик LSTM модели в таблицу
results["LSTM"] = {"RMSE": rmse, "MSE": mse, "MAE": mae, "R2": r2}

# Обновление metrics_df с новыми результатами
metrics_df = pd.DataFrame(results)
metrics_df
```

Исходя из таблицы метрик для различных моделей, можно сделать следующие выводы:

1. **Random Forest** и **Gradient Boosting** демонстрируют самые низкие значения RMSE и MSE среди всех моделей, что указывает на их относительно лучшую точность прогнозирования по сравнению с другими моделями.

2. Модели **Linear Regression** и **SVR** показывают значительно более высокие значения RMSE и MSE, а также отрицательные значения R2. Это может указывать на то, что эти модели плохо соответствуют данным и не способны хорошо объяснить изменчивость целевой переменной.

3. Модель **CatBoost** демонстрирует сравнительно хорошие значения RMSE и MSE, однако R2 оказывается ниже, чем у моделей случайного леса и градиентного бустинга.

4. Похоже, что LSTM модель показывает крайне низкую эффективность, согласно представленным метрикам. Очень высокие значения RMSE, MSE и MAE, а также отрицательное значение R2 говорят о том, что LSTM модель значительно отклоняется от реальных данных и плохо соответствует трендам.

Исходя из этих результатов, модели случайного леса и градиентного бустинга представляются наиболее подходящими для прогнозирования данного набора данных, в то время как модель LSTM следует пересмотреть или использовать с более тщательной настройкой параметров.
